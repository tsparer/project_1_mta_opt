```python
import numpy as np
import pandas as pd

#  Question 1 answer
mta_d = pd.read_csv("mta_data_scratch.txt")



##########################################################################################

# Question 2 answer:  Takes the dataframe from question 1 and turns it into a smaller data frame containing
# date/time, entries C/A, unit, SCP, Station, and Entries.

#pd.to_datetime()

  # creates new dataframe

def clean_date_time(df):
    dt  = df["DATE"] +" "+ df["TIME"]   # creates a dataframe "dt"- entries are strings of dates and times
    ndt = pd.to_datetime(dt)
    print(ndt)
    return ndt                 # converts the dates and times to datetime objects

def ans_df(df):
    print ("df for ans")
    print (df)
    date_time = clean_date_time(df)
    print ("datetime")
    print (date_time)
    
    df.drop(["DATE","TIME"], inplace = True)
    challenge_q_2_data = df[["C/A", "UNIT","SCP", "STATION", "ENTRIES",  ]]
    result = pd.concat([date_time, challenge_q_2_data])
        #adds the datetime dataframe to the answer dataframe
    print(result)
    
answer = ans_df(mta_d)

print ("no flags yet?")
print(answer)
    
    
# print(challenge_q_2_data.head())

#########################################################################################
# 3 Total daily entries
# hint use groupby



############################################################################################
# 4 plot daily time series for turnstyle

########################################################################



                    
```
